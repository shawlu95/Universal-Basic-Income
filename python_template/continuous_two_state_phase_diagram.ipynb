```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# number of arrows on each axis
n_arrows = 20
```

```python
def plot_phase(X, Y, x_dot, y_dot, figsize=(8, 8)):
    # normalize growth rate 
    M = (np.hypot(x_dot, y_dot)) 
    
    # avoid zero division errors 
    M[ M == 0] = 1. 
    
    normalize each arrows
    x_dot /= M # 
    y_dot /= M

    fig, ax = plt.subplots(figsize=figsize)
    ax.quiver(X, Y, x_dot, y_dot, M, pivot='mid')
    ax.set_aspect('equal')
    plt.show()
```

```python
# define parameter range
x = np.linspace(-2, 2, n_arrows)
y = np.linspace(-2, 2, n_arrows)

# define parameter grid
X, Y = np.meshgrid(x, y)

# gradient of change
x_dot = - 2 * X + X * X - 4 * X * Y
y_dot = - Y + Y * Y - X * Y

plot_phase(X, Y, x_dot, y_dot)
```

```python
# define parameter range
x = np.linspace(-2, 2, n_arrows)
y = np.linspace(-2, 2, n_arrows)

# define parameter grid
X, Y = np.meshgrid(x, y)

# gradient of change
x_dot = - X + Y ** 2
y_dot = - Y * (X + 1)

plot_phase(X, Y, x_dot, y_dot)
```
